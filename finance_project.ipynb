```python
from pandas_datareader import data, wb
import datetime
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import cufflinks as cf
cf.go_offline()

sns.set_style('whitegrid')
plt.rcParams['axes.xmargin'] = 0
plt.rcParams['axes.ymargin'] = 0
```

# *get the data from yahoo finance:
## pandas.pydata.org/pandas-docs/stable/remote_data.html

```python
start = datetime.datetime(2006,1,1)
end = datetime.datetime(2020,1,1)
```

```python
# Bank of America
BAC = data.DataReader('BAC', 'yahoo', start, end)
BAC
```

```python
# CititiGroup
C = data.DataReader('C','yahoo',start,end)
C
```

```python
# JPMorgan Chase
JPM = data.DataReader('JPM','yahoo',start,end)
JPM
```

```python
# Morgan Stanley
MS = data.DataReader('MS','yahoo',start,end)
MS
```

```python
# Wells Fargo
WFC = data.DataReader('WFC','yahoo',start,end)
WFC
```

```python
# Goldman Sachs
GS = data.DataReader('GS','yahoo',start,end)
GS
```

```python
# sanity check ---> all the sizes must match
assert (GS.shape == (3523,6)) & (WFC.shape == (3523,6)) & (MS.shape == (3523,6)) & (BAC.shape == (3523,6)) & (JPM.shape == (3523,6)) & (C.shape == (3523,6))
```

```python
thickers = ['BAC', 'C', 'GS', 'JPM', 'MS', 'WFC']
```

```python
# create a bank_stocks dataframe 
bank_stocks = pd.concat([BAC, C, GS, JPM, MS, WFC], axis = 1, keys = thickers)
bank_stocks
```

```python
assert bank_stocks.shape==(3523,36)
```

```python
# set the column name level
bank_stocks.columns.names = ['Bank Thicker', 'Stock Info']
```

```python
bank_stocks
```

## max close price for each bank's stock throughout the time period

```python
max_close = list(map(lambda i: (i,bank_stocks[i]['Close'].max()), thickers))
max_close
```

```python
# better solution
bank_stocks.xs(key='Close', axis=1, level='Stock Info').max()
```

## create a dataframe with all the returns for each stock

```python
returns = pd.DataFrame()

for i in thickers:
    returns[i+' Return'] = bank_stocks[i]['Close'].pct_change()

returns
```

## making a pairplot of returns dataframe

```python
# the first row is Null so everything from the second row
sns.pairplot(returns[1:])
```
